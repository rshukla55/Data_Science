```python
# Scatter plot GrLivArea and SalePrice
data = pd.concat([df_train["GrLivArea"],df_train["SalePrice"]], axis=1)
data.plot.scatter("GrLivArea", "SalePrice", ylim=(0, 800000))
```

```python
# Scatter plot TotalBsmtSF and SalePrice
data = pd.concat([df_train["TotalBsmtSF"],df_train["SalePrice"]], axis=1)
data.plot.scatter(x="TotalBsmtSF", y="SalePrice", ylim=(0, 900000))
```

```python
# box plot OverallQual/SalePrice
data = pd.concat([df_train["OverallQual"],df_train["SalePrice"]], axis=1)
f, ax = plt.subplots(figsize=(10,8))
fig = sns.boxplot(x="OverallQual", y="SalePrice", data=data)
fig.axis(ymin=0, ymax=800000)
# plt.show()
```

```python
data = pd.concat([df_train['YearBuilt'], df_train['SalePrice']], axis=1)
fig, ax = plt.subplots(figsize=(20,8))
boxplot_axis = sns.boxplot(x="YearBuilt", y="SalePrice", data=data)
boxplot_axis.axis(ymin=0, ymax=800000)
plt.xticks(rotation=90)
plt.show()
```

# Correlation

```python
corrmat = df_train.corr()
print(corrmat)

plt.subplots(figsize=(30,15))

sns.heatmap(corrmat, vmax=.8, square=True)
plt.show()
```

```python
k=10
cols= corrmat.nlargest(k, 'SalePrice')['SalePrice'].index
cm = np.corrcoef(df_train[cols].values.T)
# df_train[cols].values
plt.subplots(figsize=(20,10))
sns.set(font_scale=1.25)
sns.heatmap(cm, cbar=True, annot=True, square=True, xticklabels=cols.values, 
            yticklabels=cols.values, fmt='0.2f', annot_kws={'size':10})
plt.show()
```

```python
# scatterplot
sns.set()
print(cols.values)
sns.pairplot(df_train[cols.values], size=2.5)
plt.show()
```

# Missing values

```python
total = df_train.isnull().sum().sort_values(ascending=False)
percentage = ((df_train.isnull().sum())/(df_train.isnull().count())).sort_values(ascending=False)

missing_data = pd.concat([total, percentage], axis=1, keys=["Total", "Percent"])
missing_data.head(20)
```

```python
# dealing with missing data

print((missing_data[missing_data["Total"] > 1]).index)
df_train = df_train.drop((missing_data[missing_data["Total"] > 1]).index, 1)
df_train = df_train.drop(df_train.loc[df_train["Electrical"].isnull()].index)
```
